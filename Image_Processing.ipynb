```julia
# Generate a image from arrays(meshgrid)
N = 500
λ = 70
σ = 70
# meshgrid 
x = linspace(0,N,N)
y = x'
img_test = sin(2 * π * x / λ) .* sin(2 * π * y / λ);
```

```julia
# save the image in disk
imwrite(img_test, "Images/test.jpg")
```

```julia
# Display the image
ImageView.display(img_test)
```

Image Processing in Julia 
======================

###Basic comands 

```julia
# Load a Image Processing module
using Images
```

```julia
# In Ipython Notebook the image display automatically
im = imread("seguidor_on_off_ardublock.png")
```

```julia
# typeof Image
typeof(im)
```

```julia
# Podemos convertir a Array
imA = convert(Array, im)
```

```julia
imA_r = imA[:,:, 1]
```

```julia
imwrite(imA_r, "canal_rojo.png")
```

```julia
using Color, Images, FixedPointNumbers
```

```julia
img = imread("/home/elsuizo/images/julia_logo.jpg")
```

```julia
RGB(0,0,0)
```

```julia
m,n = size(img)
```

```julia
black = RGB(0,0,1)
color₁ = RGB(1,0,0)
tol = 0.1
color₂ = RGB(1,0+tol,0+tol)
dist = zeros(Float64,m,n)
for i in 1:m
    for j in 1:n
        dist[i,j] = colordiff(black,img[i,j])
    end
end

            
```

```julia
d₁ = colordiff(black,color₁)
```

```julia
d₂ = colordiff(black,color₂)
```

```julia
binary = (d₂ .<= dist .<= d₁)
```

```julia
using Winston
```

```julia
imagesc(binary')
```
